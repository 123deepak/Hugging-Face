```python
from huggingface_hub import HfApi

# Initialize the API
api = HfApi()

# List models with a limit of 10, sorted by downloads
models = api.list_models(
    filter="text-generation",
    limit=10,
    sort="downloads",
    full=True  # Include all model card metadata
)

# Print the model name, downloads, and size
print("Top 10 Models by Downloads:\n")
for model in models:
    # Access downloads and model size in bytes
    downloads = model.downloads

    print(f"Model: {model.modelId}, Downloads: {downloads}")
```

```python
from transformers import pipeline
model = pipeline(task='text-generation', model='openai-community/gpt2')
```

```python
# Check its parameters
# Access the model
model_info = model.model

# Calculate total number of parameters
total_params = sum(p.numel() for p in model_info.parameters())

# Calculate the number of trainable parameters
trainable_params = sum(p.numel() for p in model_info.parameters() if p.requires_grad)

print(f"Total parameters: {total_params}")
print(f"Trainable parameters: {trainable_params}")
```

```python
text = "Scientists should continue their research in science and technology because of its potential to"
```

```python
output = model(text, max_length = 75, pad_token_id = model.tokenizer.eos_token_id, truncation = True)
```

```python
print(output[0]["generated_text"])
```

```python
# Generate text with adjusted settings
output = model(
    text,
    max_length=75,  # Limit token length
    num_return_sequences=1,  # Generate one response
    do_sample=False,  # Deterministic decoding
    no_repeat_ngram_size=2,  # Avoid repeating n-grams
    pad_token_id = model.tokenizer.eos_token_id
)
```

```python
print(output[0]["generated_text"])
```
